```python
#you can define you own funciton and use automatic normalization#I know not all functions have analytic formula for normalization@normalized_function(-20,20)def tofit(x,m1,s1,m2,s2,a):    g1 = gaussian(x,m1,s1)    g2 = gaussian(x,m2,s2)    ret = a*g1+(1-a)*g2    return ret
```

```python
#see all gooduml, minu = fit_uml(tofit,twopeak,m1=0.,m2=10.,s1=2.,s2=2.,a=0.5)uml.draw(minu)
```
