```python
#lets try a function that builtin one is not availble
#this is slow though you can speed this up by using cython but for most purpose this is fast "enough"
peak1 = randn(10000)
peak2 = randn(5000)+10
twopeak = np.append(peak1,peak2)
hist(twopeak,bins=100,histtype='step');
```

```python
#you can define you own funciton and use automatic normalization
#I know not all functions have analytic formula for normalization
@normalized_function(-20,20)
def tofit(x,m1,s1,m2,s2,a):
    g1 = gaussian(x,m1,s1)
    g2 = gaussian(x,m2,s2)
    ret = a*g1+(1-a)*g2
    return ret
```

```python
%%timeit
#see all good but kinda slow
uml, minu = fit_uml(tofit,twopeak,m1=0.,m2=10.,s1=2.,s2=2.,a=.6)
uml.draw(minu)
```

```python
%%cython
cimport cython
from libc.math cimport exp,sqrt,M_PI
cdef inline double gaussian(double x,double m,double s):
    cdef double N = 1/sqrt(2*M_PI)
    return N*exp(-(x-m)*(x-m)/(s*s)/2)
@cython.binding(True)
def ctofit(double x,double m1,double s1,double m2,double s2,double a):
    cdef double g1 = gaussian(x,m1,s1)
    cdef double g2 = gaussian(x,m2,s2)
    cdef double ret = a*g1+(1-a)*g2
    return ret
```

```python
ntofit = Normalize(ctofit,(-20,20))
```

```python
%%timeit
uml, minu = fit_uml(ntofit,twopeak,m1=0.,m2=10.,s1=2.,s2=2.,a=.6)
uml.draw(minu)
```

```python
#lets try to do chi^2 regression
def f(x,a,b,c):
    return a*x**2+b*x+c
vf = vectorize(f)
x=linspace(-3,3,100)
y = vf(x,1,2,3)+randn(100)#that line + gauassian of mean 0 width 1 to give it some fluctuation
err = np.zeros(100)
err.fill(1.)
errorbar(x,y,err,fmt='.b')
xlim(-3,3)
#now we got y and x
```

```python
to_minimize = Chi2Regression(f,x,y,err)
m=RTMinuit.Minuit(to_minimize)
m.migrad()
```

```python
print m.values
print m.errors
to_minimize.draw(m)
```

```python
#fixing parameter is easy to through minuit
to_minimize = Chi2Regression(f,x,y,err)
m=RTMinuit.Minuit(to_minimize,c=3.,fix_c=True)#just that
m.migrad()
print m.values
print m.errors#yep error on c is the default one
to_minimize.draw(m)
```
